### Assumption:

## Power

Raise m to the power of n

```scheme
(define power
    (lambda (m n)
        (cond
             ((zero? n) 1)
             (else (* m (power m (- n 1))))
        )
    )
)
```

```scheme
(power 5 2)
```

```scheme
(power 100 0)
```

## Rempick

The function takes 2 arguments: n and lat, removes the element at position n from lat and returns the resulting list. It always returns a list.

```scheme
(define rempick
    (lambda (n lat)
        (cond
             ((null? lat) '())
             ((= 1 n) (cdr lat))
             (else (cons (car lat) (rempick (- n 1) (cdr lat))))
        )
    )
)
```

```scheme
(rempick 3 '(hotdogs with hot mustard))
```

## All-nums

Takes a list and returns a new list of all the numbers from the input list.

```scheme
(define all-nums
    (lambda (lat)
        (cond
             ((null? lat) '())
             ((number? (car lat)) (cons (car lat) (all-nums (cdr lat))))
             (else (all-nums (cdr lat)))
        )
    )
)
```

```scheme
(all-nums '(5 boys and 6 girls))
```

```scheme
(all-nums '(hello world!))
```

## Eqan

combine *eq?* and *=* for atoms

```scheme
(define eqan
    (lambda (a1 a2)
        (cond
             ((and (number? a1) (number? a2)) (= a1 a2))
             ((or (number? a1) (number? a2)) #f)
             (else (eq? a1 a2))
        )
    )
)
```

```scheme
(eqan 2 2)
```

```scheme
(eqan 2 'a)
```

```scheme
(eqan 'a 'a)
```

## Occur

Counts the number of times an atom appears in a list.

```scheme
(define occur
    (lambda (a lat)
        (cond
             ((null? lat) 0)
             ((eqan a (car lat)) (+ 1 (occur a (cdr lat))))
             (else (occur a (cdr lat)))
        )
    )
)
```

```scheme
(occur 4 '())
```

```scheme
(occur 5 '(525 5 hello 5 8 9))
```

```scheme
(occur 'hello '(525 hello world hello you 8 9))
```
